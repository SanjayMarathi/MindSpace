```python
import pickle
with open('./content/model_nb.pkl','rb') as f:
    model = pickle.load(f)

print("Type:", type(model))
print("Has predict_proba:", hasattr(model,'predict_proba'))
print("Has classes_:", hasattr(model,'classes_'))
if hasattr(model,'classes_'):
    print("classes_:", model.classes_)

# test sample texts
# samples = [
#     "I am feeling very sad and lonely. Nothing seems to make sense anymore.",
#     "The sky is blue and the grass is green.",
#     "I feel fine and have no problems."
# ]
samples = [
    "im feeling good  , life is awesome .  I m really happy "
]

for s in samples:
    if hasattr(model,'predict_proba'):
        p = model.predict_proba([s])[0]
        print("sample:", s[:40], "-> proba:", p)
    else:
        p = model.predict([s])[0]
        print("sample:", s[:40], "-> predict:", p)
```
